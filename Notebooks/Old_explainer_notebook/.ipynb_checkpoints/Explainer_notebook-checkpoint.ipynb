# 1. Motivation

#### The data sets
The data used in this project consists of two data sets. One contains data from the conversation on Twitter surrounding Sarah Everard, and the other contains random sampled tweets by quering the neutral pronoun _I_. Our data sets contain 15 variables after cleaning and preprocessing and to name a few it contains the user's name, the text of the tweet, geographical location of the user and the predicted gender of the user. 


#### Why Twitter and why Sarah-data?
<!--  - __Why did you choose this/these particular dataset(s)?__  -->
We have choosen Twitter as the platform from which we would like to study social interaction due to its user audience and its textbased interaction functionality. We chose to collect tweets based on the hashtag #SarahEverard to shed light on women's safety due to the case of her murder being current and a case of violence against women. We cleaned the data by removing variables not necessary for the following data analysis. We also cleaned the time data to easily process the dates and create graphs showing the timeline of the events that undfolded during the twitter activity. ## The work with text 
### Twitter API (searchtweets)
We were lucky enough to get access to the Twitter API. We recieved two developer account permissions, but unfortunatley not the academic permission, which would have given us access to historical data. We could therefore only go 30 days back on each of our data pulls. This resulted in us not being able to extract tweets from the day of the murder, but the days following the events of the murder. 

The API creates a stream of tweets from a rule. The rule to gather data related to Sarah Everard was "saraheverrad" resulting in getting tweets related to that keyword. The stream gather information of the user, time, tweet text, location and more. It also counts a retweet as a tweet, since there technically is a new author on this tweet. 

#### Limitations
The API from Twitter limits us in how many pull requests and tweets we can get each month. The limitations are set to 25.000 tweets every month and 250 pull requests, where the maximum tweets per pull request are 100 tweets. This limits our dataset significantly, since our data gathering from the first time pulling and second time pulling is within 30 days, creating an overlap in tweets pulled. We still managed to gather a lot of tweets, but this still limitied our maxmimum amount of data. 

### Gender Infering API
One of our main issues with the data was that twitter users do not inform about their gender when creating an account. Therefore, we were forced to try to infer the gender upon the users. We used a python library called M3infernece to do so. M3inference takes a jsonl (json file seperated by new lines) as a user to predict. It is perfect for Twitter users, as the library needs serveral variables to predict a gender. It needs:
> - id
> - name
> - screen_name
> - description
> - lang

It is also possible to predict upon the twitter picture, but due to time constrains, this was avoided. From this, the model predictes with percentages how sure it is in the user being a male or female. We then extracted the gender with the highest probability percentage and added that gender as a new column for the dataframe. 

### Assumptions of Location
From the extracted data, we could also retrive the location of the users. From these data, it was possible to see that the majority of our data belonged to users originating from the UK. To check against the baseline, we see a much more spread out locaion wise, which leads us to conclude that the conversation on the matter of Sarah Everad is mostly located in the UK. 

### Visulisation of the sentiment

For visualising the words used and looking at sentiment in the text we used several tools.

#### Wordclouds & Term Frequency
Wordclouds are an easy way to visualise the most common tokens used across tweets. Our token representation are based on how much a tweet has been retweeted. Therefore, the same tweet may occur multiple times, but this is because many people share that opinon and wants to spread that message. We have presented wordclouds for both men and women to look at the differences between the words used. Funny enough, the keyword "Saraheverard" is the number one most frequent word for both men and female. This is because we only extract tweets related to Everard. The Term frequency is computed from generating the text into tokens, where stopwords and pronouns are removed. We therefore only get "Terms" and not links or other unwanted "non"-words as tokens.

## The work with networks

### Creating the networks 
As said, two data sets have been collected - one containing tweets related to Sarah Everard and one containing tweets sampled random from twitter. From these data sets 8 different online social networks have been visualized as graphs, and additionally one perfectly random network has been created. 
- #1 The entire Sarah-network
- #2 The exclusively females of the Sarah-network
- #3 The exclusively males of the Sarah-network
- #4 The extended female Sarah-network
- #5 The extended male Sarah-network
- #6 The entire random network
- #7 The exclusively females of the random network
- #8 The exclusively males of the random network
- #9 The perfect random network

### The networks as graphs
All networks have been created visually with the softwarepackage NetworkX for python. The first 8 graphs have been created from the collected data and subsets hereof, while the 9th network - the perfect random network - have been created as a binomial Erdős-Rényi graph based on the same properties as the entire Sarah-network. 

An important property to note regarding the creation of the female and male networks: the _male Sarah-network_ consists of male authors and both male and female parent authors. This is done to vizualise the reality of the male interaction, where males share ideas and opinions of females and the same goes for the oppisite example. 

### Choices for the networks 
The nodes of the networks will be the users interacting in either the conversations surrounding the murder of Sarah or in random conversations. A link between two nodes is created if user A retweets user B. 

For the purpose of this project the act of retweeting is assumed to indicate a shared oppinion. This assumption is accepted based on statements such as  _"(people) look for and retweet content that is credible, intelligent and valuable to them"_ (https://ignitevisibility.com/retweet-what-makes-people-share/) and _"Within the Twitter community a 'Retweet' is a social gesture indicating endorsement of an idea"_ (https://web-strategist.com/blog/2008/11/23/retweet-the-infectious-power-of-the-word-of-mouth/). This premise of conversation and like-mindedness is the foundation for the network analysis, but the original idea was to get access to the comments of a tweet. This model is thought to create better insight in the conversation but it have not been possible to obtain access to the comments. 

This is the reason for making the graphs undirected and acyclic graphs, and the edges in the graphs are weighted proportionally to the frequency by which they are retweeted.  

### Distribution of degrees 

The measuring and plotting of degree distributions were done in order to understand the arcitecture of the networks better. This can reveal whether a network is of the same structure as most social networks, and if it infact is a small world network with the property of a power law degree distribution. 

The investigation of degree distributions revealed the random twitter data to actually not be random at all. 

Average degree, standard deviation of the degrees and the max degree were also calculated and used to accompany each other and better understand the networks. The approximately equal average degrees showed huge differnces in the variation of degrees with max degrees varying from 10.00 to 9395.00. 


### The average shortest path length

This measurement of the average shortest path length is calculated by finding the shortest path between all pairs of nodes, adding them up, and then dividing by the total number of pairs (https://users.cs.northwestern.edu/~hrb281/SocialTies.html). One of the primary characteristics of the small world networks is a short "average shortest path length". This is a value relative to the size of the system, N. The "rule of thumb" value for the average shortest path length is 6, and since the Sarah-network's value is 3.842 i.e. a value below 6, we conclude the network to be a small world network. This was the expected conclusion since the network is a empirical real world network.

As a counterpart for this conclusion we would expect the random generated network to have an "average shortest path length" larger than 6, and this is exactly the observed case with a value of 11.537.

### Cluster coefficients 

The cluster coefficient is a measure of the degree of which the nodes in a graph tend to cluster together. This measure is used to investigate how closely the nodes are connected in the subgraphs of exclusive genders and in the entire network. 

We expected the cluster coefficient of the random Twitter network to be lower than for the entire Sarah data. This was also exactly what we observed. The smaller cluster coefficient of the random-network indicates that the network is comprised of numerous weak ties, and the high clustering coefficient of the Sarah-network indicates it being a small world. Both observations are supported by the other analysis done.
